### PIECHART VISULAIZATION

```python
ot=df[['OT Days','Present Days']].sort_values(by=['OT Days'],ascending=False).head()
ot
```

```python
y=[ot['OT Days'].sum(), ot['Present Days'].sum()]
lab=["OT DAYS","PRESENT DAYS"]
mclr=["r","skyblue"]
plt.pie(y,labels=lab,startangle=80,colors=mclr)
plt.legend(title="EMPLOYEE DETAILS")
plt.show()
```
